# Modeling

```python
X_cols = ['OverallQual', 'GrLivArea', 'GarageCars', '1stFlrSF', 'FullBath', 'YearBuilt', 'YearRemodAdd',
          'ExterQual','KitchenQual','BsmtQual','Foundation' ]

train_X_final = train_X[X_cols]
val_X_final   = val_X[X_cols]

rfr_param_grid = {'n_estimators': np.arange(50, 500, 50)
              ,'random_state':[0]}
rfr = RandomForestRegressor()
rfr_cv = GridSearchCV(rfr, rfr_param_grid, cv=5)
rfr_cv.fit(train_X_final, train_y)
rfr_cv.best_params_

gbr_param_grid = {'n_estimators': np.arange(50, 500, 50)
                  ,'max_features':np.arange(3, 10)}
gbr = GradientBoostingRegressor()
gbr_cv = GridSearchCV(gbr, gbr_param_grid, cv=5)
gbr_cv.fit(train_X_final, train_y)
gbr_cv.best_params_

xgr_param_grid = {'n_estimators': np.arange(50, 500, 50)
                  }
xgr = XGBRegressor()
xgr_cv = GridSearchCV(xgr, xgr_param_grid, cv=5)
xgr_cv.fit(train_X_final, train_y)
xgr_cv.best_params_

```

```python
test_X = test[X_cols]
#test_X.info()
test_X['GarageCars'].fillna(value=0, inplace=True)
#test_X.info()
predicted_prices_rfr = rfr_cv.predict(test_X)
predicted_prices_gbr = gbr_cv.predict(test_X)
predicted_prices_xgr = xgr_cv.predict(test_X)
predicted_prices = (predicted_prices_rfr + predicted_prices_gbr + predicted_prices_xgr)/3
```

```python
print('RMSE score on train data :')
predicted_prices_rfr_val = rfr_cv.predict(val_X_final)
predicted_prices_gbr_val = gbr_cv.predict(val_X_final)
predicted_prices_xgr_val = xgr_cv.predict(val_X_final)
predicted_prices_val = (predicted_prices_rfr_val + predicted_prices_gbr_val + predicted_prices_xgr_val)/3
print(np.sqrt(mean_squared_error(val_y,predicted_prices_val)))

print('RMSE score on train data with RFR:')
print(np.sqrt(mean_squared_error(val_y,rfr_cv.predict(val_X_final))))

print('RMSE score on train data with GBR:')
print(np.sqrt(mean_squared_error(val_y,gbr_cv.predict(val_X_final))))

print('RMSE score on train data with XGR:')
print(np.sqrt(mean_squared_error(val_y,xgr_cv.predict(val_X_final))))
```
